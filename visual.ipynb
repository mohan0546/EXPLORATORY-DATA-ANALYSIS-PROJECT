```python
import seaborn as sns
sns.get_dataset_names()
```

```python
import matplotlib.pyplot as plt
tt=sns.load_dataset('titanic')
tt.head(50)
```

```python
plt.Figure(figsize=(8,7))
sns.countplot(x='class', data=tt)
plt.title('passenger count by class')
plt.show
```

```python
plt.Figure(figsize=(8,7))
sns.barplot(x='class',y='pclass', data=tt)
plt.title('passenger count by class')
plt.xlabel('class')
plt.ylabel('count')
plt.show
```

```python
plt.figure(figsize=(8,7))
sns.histplot(tt['age'],bins=20,kde=True,color='black')
plt.title('Age distribution of passenger')
plt.show()
```

```python
counts=tt['survived'].value_counts()
counts
```

```python
#smot= synthetic minority oversampling technique      # logistic regression outlier detection #oversampleing / undersampling
survival_counts=tt['survived'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(survival_counts, labels=['no survive','survive'],colors=['red','green'],autopct='%1.1f%%')
plt.title('survival rate on titanic')
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.boxplot(x='fare', data=tt, palette='Set2')
plt.title('fare distribution by passengers')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.scatterplot(x='age',y='fare',hue='survived',style='sex',data=tt,palette='rainbow')
plt.title('age vs fare')
plt.xlabel('age')
plt.ylabel('fare')
plt.legend(title='surived')
plt.show()
```

```python
numerical= tt[['survived','pclass','age','sibsp','parch','fare']].corr()

plt.Figure(figsize=(5,7))
sns.heatmap(numerical,annot=True,cmap='coolwarm',center=4)
plt.title('correlation heatmap of tt ds')  #0.75posistive corr,-0.75 is negative corr , 0.075 is no corr
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.violinplot(x='class',y='age',hue='survived',split=False,data=tt,palette='rainbow')
plt.title('age distribution by class and survival')
plt.xlabel('class')
plt.ylabel('age')
plt.legend(title='surived')
plt.show()
```

```python
sns.kdeplot(data=tt['pclass'].dropna(),color='black',fill=True)
plt.title('sibsp distribution')
plt.xlabel('fare')
plt.ylabel('density')
plt.show()
```

```python
sns.pairplot(tt)
```

```python
sns.pairplot(tt[['age','fare','class']])
```

```python
plt.figure(figsize=(5,5))
sns.swarmplot(x='class',y='age',hue='survived',data=tt,dodge=True,palette='Set2')
plt.title('age distribution by class and survival')
plt.xlabel('class')
plt.ylabel('age')
plt.legend(title='surived')
plt.show()
```

```python
g=sns.FacetGrid(tt,col='class',row='survived',margin_titles=True)
g.map(sns.histplot,'age', bins=20, color='pink')
g.fig.suptitle('Age distribution by class and survived', y=1.03)
```
